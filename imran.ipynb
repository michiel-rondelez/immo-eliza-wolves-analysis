THE FINAL FUNCTION TO REMOVE THE DISCARDED PROPERTIES

```python
import numpy as np
import pandas as pd

properties = pd.read_csv('data/raw/data.csv')


def remove_discarded_houses(properties):


    # REMOVE HOUSES WITH MISSING OR EMPTY PRICE

    properties['Price'] = properties['Price'].replace('', np.nan) # Convert empty strings in 'Price' to NaN 

    
    properties.dropna(subset=['Price'], inplace=True) # Remove rows where 'Price' is NaN or None



   # REMOVE HOUSES THAT ARE RENT OR PUBLIC SALE

    # Properties where 'Type of sale' is 'Rent' or 'Public Sale'
    discarded_properties = properties['Type of sale'].isin(['Rent', 'PublicSale'])

    # Keep only the rows that are NOT Rent or PublicSale
    properties = properties[~discarded_properties]


    
    return properties


properties = remove_discarded_houses(properties)
#properties
display(properties)


```

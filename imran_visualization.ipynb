```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
properties_data = pd.read_csv("clean_data_0.csv") #change to clean data
```

```python
# --- Gr√°fica 1: Price vs Terrace (0=no, 1=yes) ---
plt.figure(figsize=(8,5))
sns.boxplot(x='Terrace', y='Price', data=properties_data, palette='Set3')
plt.title('Price vs Terrace (0=no, 1=yes)')
plt.xlabel('Terrace')
plt.ylabel('Price')
plt.ylim(0, 500000)  # Para ver mejor los precios
plt.show()
```

The boxplot shows that homes with a terrace tend to be more expensive than those without one. This is visible because the median (the line inside each box) is noticeably higher for properties with a terrace. In other words, the central 50% of prices for homes with a terrace lies at higher values, indicating that having a terrace is associated with a higher property price.

```python
sns.scatterplot(data=properties_data, x='Terrace Surface', y='Price')
plt.title('Price vs Terrace Surface')
plt.xlabel('Terrace Surface')
plt.ylabel('Price')
#plt.ylim(0, 500000)  
#plt.xlim(0, 1000)  
plt.show()
```

```python
sns.scatterplot(data=properties_data[properties_data["Terrace Surface"] < 1000], x='Terrace Surface', y='Price', alpha=0.4)
plt.title('Price vs Terrace Surface')
plt.xlabel('Terrace Surface')
plt.ylabel('Price')
#plt.ylim(0, 500000)  
#plt.xlim(0, 1000)  
plt.show()
```

```python
# Correlation calculation 2
correlation2 = properties_data['Terrace Surface'].corr(properties_data['Price'])
print(f"Correlation between Terrace Surface and Price: {correlation2}")
```

```python
sns.scatterplot(data=properties_data[properties_data["Terrace Surface"] < 400], x='Terrace Surface', y='Price')
plt.title('Price vs Terrace Surface')
plt.xlabel('Terrace Surface')
plt.ylabel('Price')
#plt.ylim(0, 500000)  
#plt.xlim(0, 1000)  
plt.show()
```

```python
# Filtrar terrazas menores de 400
filtered_data = properties_data[properties_data["Terrace Surface"] < 400]

# Calcular la correlaci√≥n
correlation2 = filtered_data['Terrace Surface'].corr(filtered_data['Price'])
print(f"Correlation between Terrace Surface (<400) and Price: {correlation2}")
```

The original correlation was very low due to extreme terrace outliers.

By limiting terrace surface to <400 m¬≤, the correlation increases significantly,
revealing a more realistic and representative relationship between Terrace Surface and Price.


```python
sns.scatterplot(data=properties_data[properties_data["Terrace Surface"] < 400], x='Terrace Surface', y='Price',alpha=0.4)
plt.title('Price vs Terrace Surface')
plt.xlabel('Terrace Surface')
plt.ylabel('Price')
#plt.ylim(0, 500000)  
#plt.xlim(0, 1000)  
plt.show()
```
